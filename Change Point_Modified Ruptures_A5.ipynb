```python
#modified ruptures by KNJ
import matplotlib.pyplot as plt
import ruptures as rpt

## read data from A5 exercise

n_samples, dim, sigma = 40, 2, 1  #dim was 3, sigma assumed as 1
with open('A5 Exercise_test641data.txt') as newfile:
    signal = newfile.read()
n_bkps = 1 # number of breakpoints to find before stopping
signal, bkps = rpt.pw_constant(n_samples, dim, n_bkps, noise_std=sigma)

# detection
algo = rpt.Pelt(model="rbf",min_size=1, jump=1).fit(signal)  #added 'min_size=1, jump=3'
result = algo.predict(pen=1) #pen was = 1, changed due to smaller data values

# display
rpt.display(signal, bkps, result) 
plt.show()
```
