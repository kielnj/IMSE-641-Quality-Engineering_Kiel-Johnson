```python
#modified ruptures by KNJ
import matplotlib.pyplot as plt
import ruptures as rpt

## read data from Sullivan_xi excel file that has the data from the paper

n_samples, dim, sigma = 60, 2, 1.23  #dim was 3, sigma assumed as 1.23 per the paper
with open('Sullivan_xi.txt') as newfile:
    signal = newfile.read()
n_bkps = 4 # number of breakpoints to find before stopping
signal, bkps = rpt.pw_constant(n_samples, dim, n_bkps, noise_std=sigma)

# detection
algo = rpt.Pelt(model="rbf",min_size=2, jump=1).fit(signal)  #added 'min_size=2, jump=1'
result = algo.predict(pen=1) #pen was = 1, changed due to smaller data values

# display
rpt.display(signal, bkps, result) 
plt.show()
```
